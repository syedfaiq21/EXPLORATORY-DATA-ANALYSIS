# 9. Investment Patterns :Which shark invests the most frequently, and in which types of businesses?

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have your data loaded into a DataFrame called 'df'
# Replace 'ashneer_deal' and 'ghazal_deal' with the column names you want to plot
columns_to_plot = ['ashneer_deal','anupam_deal','aman_deal','namita_deal','vineeta_deal','peyush_deal','ghazal_deal']

# Create subplots for each column
fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_plot), figsize=(12, 5))

# Plot histograms for each column
for i, column in enumerate(columns_to_plot):
    axes[i].hist(df[column], bins=[-0.5, 0.5, 1.5], align='mid', rwidth=0.8, edgecolor='k')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'{column}')
    axes[i].set_xticks([0, 1])
    axes[i].set_xticklabels(['No Deal', 'Deal'])

plt.tight_layout()
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
from scipy import stats

# Load the dataset
df = pd.read_csv('Shark Tank India Dataset.csv')

# Extract the column of interest
deal_amount_column = df['deal_amount']

# Specify the expected mean under the null hypothesis
expected_mean = 50000  # Replace with the appropriate value

# Perform the one-sample t-test
result = stats.ttest_1samp(deal_amount_column, expected_mean)

# Interpret the results
alpha = 0.05
if result.pvalue < alpha:
    print(f"Reject the null hypothesis. There is a significant difference from the expected mean.")
else:
    print("Fail to reject the null hypothesis. No significant difference from the expected mean.")

# Optionally, print the t-statistic and p-value
print(f"T-statistic: {result.statistic}")
print(f"P-value: {result.pvalue}")
```
